### 9. 이후 Price를 예측하는 모델을 제작하기 위해 추가적인 변수를 만드려고 합니다. Price의 데이터 분포를 확인하고 1개 이상의 EDA를 통해 의견을 제시해주세요.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

plt.rc('font', family='malgun gothic')
plt.rc('axes', unicode_minus=False)
```

### Target 변수 - Price 분포확인

```python
sns.distplot(train['price'])
```

-> 왼쪽으로 치우친 분포이므로, log 를 취해주면 아래와 같이 종모양의 분포(정규분포에 더 가까운)를 띈다.

```python
train['price'] = np.log1p(train['price'])
sns.distplot(train['price'])
```

### 설명변수들 분포확인

```python
categorical = ['bedrooms', 'bathrooms','grade','view','waterfront','condition','floors']
continuous=['sqft_living','sqft_lot','sqft_above', 'sqft_basement', 'yr_built','sqft_living15', 'sqft_lot15']
etc = ['id','zipcode','date','yr_renovated','year', 'month', 'day','lat', 'long']
```

```python
train[continuous].hist(bins=100,figsize=(30,25))
plt.show()
```

```python
train[categorical].hist(figsize=(30,25))
plt.show()
```

### Price와의 상관관계확인

```python
#price correlation matrix
k = 20
corrmat = train.corr()
cols = corrmat.nlargest(k, 'price')['price'].index
cm = np.corrcoef(train[cols].values.T)
plt.figure(figsize=(10,10))
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

### 연속형 변수들 - price 와의 plot

```python
for i in range(len(continuous)):
    plt.title(continuous[i])
    plt.scatter(train[continuous[i]],train['price'])
    plt.show()
```

### categorical 변수들 - price와의 plot

```python
fig, ax = plt.subplots(3, 3, figsize=(30, 25))
for var, subplot in zip(categorical, ax.flatten()):
    sns.boxplot(x=var, y='price', data=train, ax=subplot)
```
